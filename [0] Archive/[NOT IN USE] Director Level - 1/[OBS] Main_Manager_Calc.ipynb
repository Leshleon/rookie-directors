```python
def PrevLastServed(df, P ,S ,A):
    x = df.loc[ ((df['Person Code'] == P) & (df['Symbol'] == S) & (df['AsOnDate'] < A))]['AsOnDate'].max()
    return x
```

```python
dff = dff.sort_values(by = ['Person Code','Symbol', 'AsOnDate'], axis = 0)
dff.loc[:,'PrevLastServed'] = dff.progress_apply( lambda x: PrevLastServed(dff, x['Person Code'], x['Symbol'], x['AsOnDate']), axis = 1)
```
