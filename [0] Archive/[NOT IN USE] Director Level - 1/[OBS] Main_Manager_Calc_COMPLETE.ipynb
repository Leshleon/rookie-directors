```python
import pandas as pd
```

```python
df1 = pd.read_csv("Main_Manager_Calc v280824.1.csv")
```

```python
df2 = pd.read_excel("Main.xlsx")
```

```python
df2['AsOnDate'] = pd.to_datetime(df2['AsOnDate'], format = '%Y-%m-%d')
df1['AsOnDate'] = pd.to_datetime(df1['AsOnDate'], format = '%Y-%m-%d')

df2['Date of Birth'] = pd.to_datetime(df2['Date of Birth'], format = '%d-%b-%Y')

df2['Appointment Date'] = pd.to_datetime(df2['Appointment Date'], format = '%d-%b-%Y')

df2['Cessation Date'] = pd.to_datetime(df2['Cessation Date'], format = '%d-%b-%Y', errors = 'coerce')

df2['Cessation Reason'] = df2['Cessation Reason'].astype(str)
```

```python
dff1 = df1.set_index(['Symbol', 'Person Code', 'AsOnDate'], drop = True)
dff1
```

```python
dff2 = df2.set_index(['Symbol', 'Person Code', 'AsOnDate'], drop = True)
dff2
```

```python
dfs = dff1.merge(dff2, on = ['Symbol', 'Person Code', 'AsOnDate'], how = 'outer', validate = '1:1')
```

```python
dfs.to_csv("Main_Manager_Calc_COMPLETE v280824.1.csv")
```
