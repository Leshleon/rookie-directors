```python
dfes.loc[dfes['Independent (Yes/No)'] == 'YES', 'IsIndep'] = 'Y'
dfes.loc[dfes['Independent (Yes/No)'] == 'NO', 'IsIndep'] = 'N'
dfes = dfes.drop('Independent (Yes/No)', axis = 1)
dfes
```

```python
dfes['PrevWORK'] = dfes.AsOnDate.shift()
dfes.insert(2, 'PrevLastServed', np.where( (dfes.AsOnDate <= dfes.PrevWORK) | (pd.isnull(dfes.PrevWORK) == True) , pd.NaT , dfes.PrevWORK))
dfes.PrevLastServed = pd.to_datetime(dfes.PrevLastServed)
dfes.drop('PrevWORK', axis = 1, inplace = True)
```

```python
dfes.insert(6,'CloBalUnc', 0)
dfes.insert(7,'CloBalIndep', 0)
dfes.insert(8,'CloBalNonIndep', 0)
dfes
```

```python
dfes
```

```python
dfes.insert(6,'cumsum_U', 0)
dfes.insert(6,'cumsum_Y', 0)
dfes.insert(6,'cumsum_N', 0)
dfes
```

```python
dfes.loc[ dfes['IsIndep'] == 'U','cumsum_U'] = 1
dfes.loc[ dfes['IsIndep'] == 'Y','cumsum_Y'] = 1
dfes.loc[ dfes['IsIndep'] == 'N','cumsum_N'] = 1
```

```python
dfes
```

```python
def OpBal1(frame):
    frame['CloBalUnc'] += (frame.loc[ frame['IsIndep'] == 'U']['AsOnDate'].min() - frame['Appointment Date']).dt.days

    frame['cumsum_U'] = np.cumsum(frame['cumsum_U'])
    frame['cumsum_Y'] = np.cumsum(frame['cumsum_Y'])
    frame['cumsum_N'] = np.cumsum(frame['cumsum_N'])
    return (frame)
```

```python
def OpBal2(frame):
    
    frame.loc[ (frame['cumsum_U'] > 0 ) , 'CloBalUnc' ] += (frame['cumsum_U']) * 365
    
    
    frame.loc[ (frame['cumsum_Y'] > 0 ) , 'CloBalIndep' ] += (frame['cumsum_Y']) * 365
    
    
    frame.loc[ (frame['cumsum_N'] > 0 ) , 'CloBalNonIndep' ] += (frame['cumsum_N']) * 365
        
    return (frame) 
```

```python
check1 = dfes.reset_index(drop = True).groupby(['Person Code', 'Symbol'] ).progress_apply(OpBal1)
```

```python
check2 = check1.reset_index(drop = True).groupby(['Person Code', 'Symbol'] ).progress_apply(OpBal2)
```

```python
check1
```

```python
check2['CloBalTotal'] = 0
check2['CloBalTotal'] = check2['CloBalUnc'] + check2['CloBalIndep'] + check2['CloBalNonIndep']
```

```python
check2.insert(13, 'OpBalUnc', 0)
check2.insert(14, 'OpBalIndep', 0)
check2.insert(15, 'OpBalNonIndep', 0)
check2.insert(16, 'OpBalTotal', 0)

check2['OpBalUnc'] = check2['CloBalUnc']
check2['OpBalIndep'] = check2['CloBalIndep']
check2['OpBalNonIndep'] = check2['CloBalNonIndep']

check2.loc[ (check2['IsIndep'] == 'U'),'OpBalUnc'] -= (365)
check2.loc[ check2['IsIndep'] == 'Y','OpBalIndep'] -= (365)
check2.loc[ check2['IsIndep'] == 'N','OpBalNonIndep'] -= (365)
check2['OpBalTotal'] = 0
check2['OpBalTotal'] = check2['OpBalUnc'] + check2['OpBalIndep'] + check2['OpBalNonIndep']
```
