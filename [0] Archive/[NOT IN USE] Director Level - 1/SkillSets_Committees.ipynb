```python
dfsc = dfcc.merge(dfms, how = 'left').dropna(subset = ['Committee Name', 'Skills'], axis = 0)
dfsc
# COMMITTEE + SKILL with inner join
```

```python
#Let's map this bitch

c = dfsc.drop(['AsOnDate', 'Skills', 'Position', 'Symbol'], axis = 1).drop_duplicates().reset_index(drop = True)
c
```

```python
d = dfms
d
```

```python
merged_df = pd.merge(c, d, on='Person Code')
merged_df
```

```python
unique_skills = set(skill for sublist in merged_df['Skills'] for skill in sublist)
unique_skills
```

```python
matrix = pd.DataFrame(0, index=merged_df['Committee Name'].unique(), columns=list(unique_skills))
matrix
```

```python
for _, row in merged_df.iterrows():
    committee = row['Committee Name']
    for skill in row['Skills']:
        if skill in matrix.columns:
            matrix.at[committee, skill] += 1

# Display the resulting matrix
print(matrix)
```
