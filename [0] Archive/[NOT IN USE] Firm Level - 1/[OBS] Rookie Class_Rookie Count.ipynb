```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv('Main_Manager_Calc v280824.1.csv')
```

```python
df1
```

```python
df2 = df1.set_index(['Person Code', 'AsOnDate', 'Symbol'], drop = False)
df2.insert(len(df2.columns), column = 'YOE', value = 0)
df2
```

```python
df2['AsOnDate'] = pd.to_datetime(df2['AsOnDate'], format = '%Y-%m-%d')

df2['Appointment Date'] = pd.to_datetime(df2['Appointment Date'], format = '%Y-%m-%d')
```

```python
df2['YOE'] = (df2['AsOnDate'] - df2['Appointment Date']).dt.days
```

```python
df2
```

```python
def YOE(frame):
    return frame['YOE'].max()
```

```python
YOEmax = df2.groupby(level = ['Person Code', 'AsOnDate']).progress_apply(YOE)
```

```python
YOEmax = YOEmax.reset_index()
YOEmax
```

```python
df2['AsOnDate'] = pd.to_datetime(df2['AsOnDate'], format = '%Y-%m-%d')
YOEmax['AsOnDate'] = pd.to_datetime(YOEmax['AsOnDate'], format = '%Y-%m-%d')


df3 = df2.reset_index(drop = True).merge(YOEmax, on = ['Person Code', 'AsOnDate'], how = 'left')
df3 = df3.rename(columns = {0 : "YOEmax"})
df3
```

```python
df3.loc[ df3['YOEmax'] <= 365*3, 'IsRookie'] = 'Y'
df3.loc[ df3['YOEmax'] > 365*3, 'IsRookie'] = 'N'

df3
```

```python
frame2 = pd.DataFrame(columns = ['Symbol', 'AsOnDate', 'Total', 'Rookie'])
frame2
```

```python
def RookieCount(frame):
    frame2 = pd.DataFrame()
    frame2['Symbol'] = frame['Symbol']
    frame2['AsOnDate'] = frame['AsOnDate']
    frame2['Total'] = frame['Person Code'].count()
    frame2['Rookie'] = frame.loc[ frame['IsRookie'] == 'Y']['Person Code'].count()
    return frame2
```

```python
FL = df3.set_index(['Symbol', 'AsOnDate'], drop = False).groupby( level = ['Symbol','AsOnDate'] ).progress_apply(RookieCount)
```

```python
FL = FL.reset_index(drop = True).drop_duplicates().reset_index(drop = True)
```

```python
df3.to_csv("Main_Manager_Calc_n_Rookie Classification v280824.1.csv")
FL.to_csv("Main_Firm_Rookie Count v280824.1.csv")
```
