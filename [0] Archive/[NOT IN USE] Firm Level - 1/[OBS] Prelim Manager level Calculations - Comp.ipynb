```python
def PastDirect(frame):
    frame3 = pd.DataFrame(index = frame.index, columns = ['AsOnDate','Past Directorships'])
    
    for i in range(len(frame)):
        Date = frame.iloc[i]['AsOnDate']
        filtered_frame = frame.loc[ frame['AsOnDate'] < Date , ['Company', 'IsIndep']]
        past_directors = list(zip(filtered_frame['Company'], filtered_frame['IsIndep']))
        past_dir = list(set(past_directors))

        frame3.at[ frame.index[i], 'Past Directorships'] = past_dir
        frame3.at[ frame.index[i], 'AsOnDate'] = Date

    return frame3
```

```python
def CurrentDirect2(group, cols):
    # Ensure cols is a list of column names and that they exist in the DataFrame
    if not all(col in group.columns for col in cols):
        raise ValueError("One or more columns in 'cols' are not in the DataFrame")

    # Create unique list of tuples from 'Company' and 'IsIndep' in the current group
    comp_stat = list(set(zip(group.Company, group.IsIndep)))
    
    # Initialize the new DataFrame
    
    frame2 = pd.DataFrame(index=group.index, columns=['Current Direct'])
    
    # Populate 'Current Direct' column
    for i in range(len(group)):
        # Get the value for each column
        row_values = [str(group[col].iloc[i]) for col in cols]
        # Check if any company from comp_stat is in row_values
        current_direct = [
            item for item in comp_stat if item[0] in row_values
        ]
        frame2.loc[group.index[i], 'Current Direct'] = current_direct
    
    return frame2
```
