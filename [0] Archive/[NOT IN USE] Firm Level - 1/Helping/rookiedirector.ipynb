```python
def getCity(row):
    rv = ""
    df = dflocations[(dflocations['lo']== row.Symbol) | (dflocations['Company'] == row.Company)]
    rv = df['Final'].iloc[0]
    return rv

def getTotalDirectors(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1))]
    rv = df.shape[0]
    return rv;

def getTotalDirectorsFirms(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1))]
    rv = len(set(df['Symbol']))
    return rv;

def getFirstYearDirectors(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1)) & (dfdirector['OpBalTotal'].isin(range(0,13))) ]
    rv = df.shape[0]
    return rv;

def getFirstYearDirectorsFirms(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1)) & (dfdirector['OpBalTotal'].isin(range(0,13))) ]
    rv = len(set(df['Symbol']))
    return rv;

def getFirstYearIndependentDirectors(row):
    rv = 0
    df = dfdirector[(dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1)) & (dfdirector['OpBalIndep'].isin(range(0,13)))]
    rv = df.shape[0]
    return rv;

def getFirstYearIndependentDirectorsFirms(row):
    rv = 0
    df = dfdirector[(dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1)) & (dfdirector['OpBalIndep'].isin(range(0,13))) ]
    rv = len(set(df['Symbol']))
    return rv;

def getTotalIndependentDirectors(row):
    rv = 0
    df = dfdirector[(dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1))]
    rv = df.shape[0]
    return rv;

def getTotalInependentDirectorsFirms(row):
    rv = 0
    df = dfdirector[ (dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & (dfdirector['AsOnYear'] == (row.AsOnYear-1))]
    rv = len(set(df['Symbol']))
    return rv;

def getDirectorsWithoutDirectorShip1Year(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( (dfdirector['AsOnYear']-row.AsOnYear == 0)) & (dfdirector['OpBalTotal'] == 0) ]
    rv = len(set(df['DirCompId']));
    return rv;
def getDirectorsWithoutDirectorShip1YearFirm(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( (dfdirector['AsOnYear']-row.AsOnYear == 0)) & (dfdirector['OpBalTotal'] == 0) ]
    rv = len(set(df['Symbol']))
    return rv;
def getDirectorsWithoutDirectorShip3Year(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( dfdirector['AsOnYear'].isin([row.AsOnYear, row.AsOnYear-1, row.AsOnYear-2]) )  & (dfdirector['OpBalTotal'] == 0) ]
    #rv = df.shape[0]
    rv = len(set(df['DirCompId']));
    return rv;
def getDirectorsWithoutDirectorShip3YearFirms(row):
    rv = 0
    df = dfdirector[(dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( dfdirector['AsOnYear'].isin([row.AsOnYear, row.AsOnYear-1, row.AsOnYear-2]) )  & (dfdirector['OpBalTotal'] == 0) ]
    rv = df.shape[0]
    rv = len(set(df['Symbol']));
    return rv;
def getIndependentDirectorsWithoutDirectorShip1Year(row):
    rv = 0
    df = dfdirector[((dfdirector['IsIndep'] == 1) & dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( (dfdirector['AsOnYear']-row.AsOnYear == 0)) & (dfdirector['OpBalIndep'] == 0) ]
    rv = len(set(df['DirCompId']));
    return rv;
def getIndependentDirectorsWithoutDirectorShip1YearFirm(row):
    rv = 0
    df = dfdirector[(dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( (dfdirector['AsOnYear']-row.AsOnYear == 0)) & (dfdirector['OpBalIndep'] == 0) ]
    rv = len(set(df['Symbol']))
    return rv;
def getIndepndentDirectorsWithoutDirectorShip3Year(row):
    rv = 0
    df = dfdirector[( (dfdirector['IsIndep'] == 1) & dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( dfdirector['AsOnYear'].isin([row.AsOnYear, row.AsOnYear-1, row.AsOnYear-2]) )  & (dfdirector['OpBalIndep'] == 0) ]
    #rv = df.shape[0]
    rv = len(set(df['DirCompId']));
    return rv;
def getIndepndentDirectorsWithoutDirectorShip3YearFirms(row):
    rv = 0
    df = dfdirector[ (dfdirector['IsIndep'] == 1) & (dfdirector['Location'] == row.Location)  & (dfdirector['Symbol'] != row.Symbol)  & ( dfdirector['AsOnYear'].isin([row.AsOnYear, row.AsOnYear-1, row.AsOnYear-2]) )  & (dfdirector['OpBalIndep'] == 0) ]
    rv = df.shape[0]
    rv = len(set(df['Symbol']));
    return rv;
```

```python
dffirm['Location'] = dffirm.apply(getCity,axis = 1)
dfdirector['Location'] = dfdirector.apply(getCity, axis =1 )

dffirm['TotDirectorsInCityT-1'] = dffirm.apply(getTotalDirectors,axis = 1)
dffirm['TotDirectorsInCityT-1Firms'] = dffirm.apply(getTotalDirectorsFirms,axis = 1)
dffirm['FirstYearDirectorsInCityT-1'] = dffirm.apply(getFirstYearDirectors,axis = 1)
dffirm['FirstYearDirectorsInCityT-1Firms'] = dffirm.apply(getFirstYearDirectorsFirms,axis = 1)

dffirm['FirstYearIndependentDirectorsInCityT-1'] = dffirm.apply(getFirstYearIndependentDirectors,axis = 1)
dffirm['FirstYearIndependentDirectorsInCityT-1Firms'] = dffirm.apply(getFirstYearIndependentDirectorsFirms,axis = 1)
dffirm['TotFirstYearIndependentDirectorsInCityT-1'] = dffirm.apply(getFirstYearIndependentDirectors,axis = 1)
dffirm['TotFirstYearIndependentDirectorsInCityT-1Firms'] = dffirm.apply(getFirstYearIndependentDirectorsFirms,axis = 1)

dffirm['DirectorsWithout1year'] = dffirm.apply(getDirectorsWithoutDirectorShip1Year,axis = 1)
dffirm['DirectorsWithout1yearFirms'] = dffirm.apply(getDirectorsWithoutDirectorShip1YearFirm,axis = 1)
dffirm['DirectorsWithout3year'] = dffirm.apply(getDirectorsWithoutDirectorShip3Year,axis = 1)
dffirm['DirectorsWithout3yearFirms'] = dffirm.apply(getDirectorsWithoutDirectorShip3YearFirms,axis = 1)

dffirm['IndependentDirectorsWithout1year'] = dffirm.apply(getIndependentDirectorsWithoutDirectorShip1Year,axis = 1)
dffirm['IndependentDirectorsWithout1yearFirms'] = dffirm.apply(getIndependentDirectorsWithoutDirectorShip1YearFirm,axis = 1)
dffirm['IndependentDirectorsWithout3year'] = dffirm.apply(getIndependentDirectorsWithoutDirectorShip3Year,axis = 1)
dffirm['IndependentDirectorsWithout3yearFirms'] = dffirm.apply(getIndependentDirectorsWithoutDirectorShip3YearFirms,axis = 1)

```
