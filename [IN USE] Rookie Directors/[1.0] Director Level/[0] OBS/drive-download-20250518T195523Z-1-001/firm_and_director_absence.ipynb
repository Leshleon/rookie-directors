```python
import pandas as pd
import glob

main_df = pd.read_csv('Main_Firm_Fin_Loc_COMPLETE v170325.csv')
main_df['AsOnDate'] = pd.to_datetime(main_df['AsOnDate'])

absence_data = []
for year in range(2015, 2025):
    file_path = f'{year} avg meetings absent %.xlsx'
    
    try:
        year_df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Could not read {file_path}: {e}")
        continue
    year_df['AsOnDate'] = pd.to_datetime(year_df['AsOnDate'])
    absence_data.append(year_df[['Symbol', 'AsOnDate', 'Average Absence (%)']])
absence_df = pd.concat(absence_data, ignore_index=True)

merged_df = pd.merge(main_df, absence_df, on=['Symbol', 'AsOnDate'], how='left')
merged_df.to_csv('Main_Firm_Fin_Loc_with_Absence.csv', index=False)
```

```python
import pandas as pd
main_df = pd.read_csv('Main_Director_COMPLETE v170325.csv')
main_df['AsOnDate'] = pd.to_datetime(main_df['AsOnDate'])
merged_df = main_df.copy()
for year in range(2015, 2025):
    file_path = f'Indian Boards Data as on 31-Mar-{year}.xlsx'
    
    try:
        year_df = pd.read_excel(file_path)
        print(f"Columns in {file_path}: {year_df.columns}")  
        
        if 'AsOnDate' not in year_df.columns:
            print(f"Warning: 'AsOnDate' not found in {file_path}. Checking for other date columns.")
            print(year_df.columns)  
            continue  
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        continue
    
    year_df['AsOnDate'] = pd.to_datetime(year_df['AsOnDate'], errors='coerce') 

    year_df = year_df[['Person Code', 'AsOnDate', 'Meetings absent (%)']]
    merged_df = pd.merge(merged_df, year_df, on=['Person Code', 'AsOnDate'], how='left')
```
