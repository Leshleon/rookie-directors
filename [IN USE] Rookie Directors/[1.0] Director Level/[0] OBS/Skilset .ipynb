```python
import pandas as pd

df = pd.read_csv('Main_Director_COMPLETE_With_Meetings.csv')
print("Before Mining")
total_rows = len(df)
print(f"Total number of rows: {total_rows}")

empty_brief_profiles = df['Brief Profile'].isna().sum() + (df['Brief Profile'].str.strip() == '').sum()
print(f"Number of empty Brief Profiles: {empty_brief_profiles}")

empty_skills = df['Skills/Competencies'].apply(
    lambda x: (isinstance(x, list) and len(x) == 0) or (not isinstance(x, list) and pd.isna(x))
).sum()
print(f"Number of empty Skills/Competencies: {empty_skills}")

empty_brief_profile_flag = df['Brief Profile'].isna() | (df['Brief Profile'].str.strip() == '')
empty_skills_flag = df['Skills/Competencies'].apply(
    lambda x: (isinstance(x, list) and len(x) == 0) or (not isinstance(x, list) and pd.isna(x))
)
intersection_count = (empty_brief_profile_flag & empty_skills_flag).sum()
print(f"Number of directors with BOTH empty Brief Profile and empty Skills/Competencies: {intersection_count}")
```

```python
skill_list = [
    'economics', 'marketing', 'fast moving consumer goods', 'accounts/audit/taxation',
    'human resources', 'healthcare', 'public policy', 'corporate social responsibility',
    'finance', 'services', 'legal/compliance', 'information technology',
    'corporate governance', 'telecommunication', 'research & development', 'strategy',
    'financial services', 'risk management', 'diversified', 'supply chain management',
    'leadership/management', 'international business', 'diversity', 'consumer discretionary',
    'commodities', 'personal values', 'utilities', 'energy', 'operations', 'industrials'
]

skill_list_lower = [skill.lower().strip() for skill in skill_list]

def extract_skills(profile):
    if pd.isnull(profile):
        return []
    profile = str(profile).lower().strip()
    matched_skills = [skill for skill in skill_list_lower if skill in profile]
    return matched_skills

df['Skills/Competencies'] = df.apply(
    lambda row: extract_skills(row['Brief Profile']) if pd.isnull(row['Skills/Competencies']) or row['Skills/Competencies'] == '' else row['Skills/Competencies'],
    axis=1
)

df['Skills/Competencies'] = df['Skills/Competencies'].apply(
    lambda x: [] if not isinstance(x, list) and pd.isnull(x) else x
)
```

```python
print("After Mining")

empty_skills = df['Skills/Competencies'].apply(lambda x: len(x) == 0).sum()
print(f"Number of empty Skills/Competencies: {empty_skills}")
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

df.reset_index(drop=True, inplace=True)
```

```python
def extract_manufacturing(nic_code):
    if pd.isnull(nic_code) or str(nic_code).strip() == '':
        return ''
    try:
        nic_str = str(int(float(nic_code))).zfill(2) 
        nic2 = int(nic_str[:2]) 
        return 1 if 10 <= nic2 <= 33 else 0 
    except:
        return ''
df['Manufacturing'] = df['NIC code'].apply(extract_manufacturing)
```

```python
import ast

def parse_nic_set(value):
    try:
        result = ast.literal_eval(value)
        return result if isinstance(result, set) else set()
    except:
        return set()

df2['AllPastDirectNIC'] = df2['AllPastDirectNIC'].apply(parse_nic_set)

df2['NIC code'] = df2['NIC code'].astype(str)
def check_same_industry(row):
    focal_nic = row['NIC code']
    past_nics = row['AllPastDirectNIC']
    
    filtered_past_nics = {str(nic) for nic in past_nics if str(nic) != focal_nic}

    past_prefixes = {nic[:2] for nic in filtered_past_nics}

    return 1 if focal_nic[:2] in past_prefixes else 0

df2['Company Business'] = df2.apply(check_same_industry, axis=1)
```

```python
import pandas as pd

committee_df = pd.read_excel(
    'Indian Boards Data as on 12-Feb-2025.xlsx',
    sheet_name='Committee Details',
    skiprows=1
)
sustainability_keywords = [
    "sustainability",
    "sustainable",
    "csr",
    "corporate social responsibility",
    "sustainability committee",
    "esg committee",
    "csr committee",
    "corporate social responsibility committee",
    "environment committee"
]
sustainability_keywords_lower = [kw.lower() for kw in sustainability_keywords]

committee_df['committee_sustainability'] = committee_df['Committee Name'].astype(str).str.lower().apply(
    lambda x: int(any(keyword in x for keyword in sustainability_keywords_lower))
)

committee_flags = (
    committee_df.groupby(['Symbol', 'Person Code'])['committee_sustainability']
    .max()
    .reset_index()
)

director_df = pd.read_csv('Main_Director_COMPLETE_Skillset.csv')

director_df['profile_sustainability'] = director_df['Brief Profile'].astype(str).str.lower().apply(
    lambda x: int(any(keyword in x for keyword in sustainability_keywords_lower))
)

merged_df = director_df.merge(
    committee_flags,
    on=['Symbol', 'Person Code'],
    how='left'
)
merged_df['committee_sustainability'] = merged_df['committee_sustainability'].fillna(0).astype(int)
merged_df['Combined_Sustainability'] = merged_df[['committee_sustainability', 'profile_sustainability']].max(axis=1)
merged_df.drop(columns=['committee_sustainability', 'profile_sustainability'], inplace=True)
```
