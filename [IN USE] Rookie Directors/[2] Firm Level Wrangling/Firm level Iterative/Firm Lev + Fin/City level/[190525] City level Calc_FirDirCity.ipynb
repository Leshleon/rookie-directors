```python
def FirDirCitymean(frame):
    frameOut = pd.DataFrame()

    zeroYearDirCityIndepMean = frame["ZeroYearDirCityIndep"].mean()
    
    zeroYearDirCityMean = frame["ZeroYearDirCity"].mean()
    
    zeroYearDirCityIndepExclMean =  frame["ZeroYearDirCityIndepExcl"].mean()
    
    zeroYearDirCityExclMean =  frame["ZeroYearDirCityExcl"].mean()
    
    
    firstYearDirCityIndepMean = frame["FirstYearDirCityIndep"].mean()
    
    firstYearDirCityMean = frame["FirstYearDirCity"].mean()
    
    firstYearDirCityIndepExclMean =  frame["FirstYearDirCityIndepExcl"].mean()
    
    firstYearDirCityExclMean =  frame["FirstYearDirCityExcl"].mean()

    
    twoYearDirCityIndepMean = frame["TwoYearDirCityIndep"].mean()
    
    twoYearDirCityMean = frame["TwoYearDirCity"].mean()
    
    twoYearDirCityIndepExclMean =  frame["TwoYearDirCityIndepExcl"].mean()
    
    twoYearDirCityExclMean =  frame["TwoYearDirCityExcl"].mean()

    
    threeYearDirCityIndepMean = frame["ThreeYearDirCityIndep"].mean()
    
    threeYearDirCityMean = frame["ThreeYearDirCity"].mean()
    
    threeYearDirCityIndepExclMean =  frame["ThreeYearDirCityIndepExcl"].mean()
    
    threeYearDirCityExclMean =  frame["ThreeYearDirCityExcl"].mean()


    frame["ZeroYearDirCityIndepMean"] = zeroYearDirCityIndepMean
    
    frame["ZeroYearDirCityMean"] = zeroYearDirCityMean
    
    frame["ZeroYearDirCityIndepExclMean"] =  zeroYearDirCityIndepExclMean
    
    frame["ZeroYearDirCityExclMean"] =  zeroYearDirCityExclMean
    
    
    frame["FirstYearDirCityIndepMean"] = firstYearDirCityIndepMean
    
    frame["FirstYearDirCityMean"] = firstYearDirCityMean
    
    frame["FirstYearDirCityIndepExclMean"] =  firstYearDirCityIndepExclMean
    
    frame["FirstYearDirCityExclMean"] =  firstYearDirCityExclMean

    
    frame["TwoYearDirCityIndepMean"] = twoYearDirCityIndepMean
    
    frame["TwoYearDirCityMean"] = twoYearDirCityMean
    
    frame["TwoYearDirCityIndepExclMean"] =  twoYearDirCityIndepExclMean
    
    frame["TwoYearDirCityExclMean"] =  twoYearDirCityExclMean

    
    frame["ThreeYearDirCityIndepMean"] = threeYearDirCityIndepMean
    
    frame["ThreeYearDirCityMean"] = threeYearDirCityMean
    
    frame["ThreeYearDirCityIndepExclMean"] =  threeYearDirCityIndepExclMean
    
    frame["ThreeYearDirCityExclMean"] =  threeYearDirCityExclMean

    return frame
```
